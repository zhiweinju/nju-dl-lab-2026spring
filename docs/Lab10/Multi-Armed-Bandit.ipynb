## 2. epsilon-贪婪算法

epsilon-贪婪算法以概率 `epsilon` 随机探索，以概率 `1 - epsilon` 选择当前估计奖励最高的拉杆。每次获得奖励后，用增量式公式更新该拉杆的奖励估计：

$$Q_k \leftarrow Q_k + \frac{1}{N_k}(r-Q_k)$$

```python
class EpsilonGreedy:
    def __init__(self, bandit, epsilon=0.01, init_prob=1.0):
        self.bandit = bandit
        self.epsilon = epsilon
        self.estimates = np.array([init_prob] * self.bandit.K, dtype=np.float64)
        self.counts = np.zeros(self.bandit.K)
        self.regret = 0.0
        self.regrets = []

    def update_regret(self, k):
        # TODO: 根据最优获奖概率和本次选择拉杆的获奖概率，更新累积懊悔
        raise NotImplementedError

    def run_one_step(self):
        # TODO: 以 epsilon 的概率随机选择拉杆，否则选择 estimates 最大的拉杆
        k = None
        r = self.bandit.step(k)
        self.counts[k] += 1
        # TODO: 使用增量式公式更新第 k 根拉杆的奖励估计
        self.update_regret(k)

    def run(self, num_steps):
        for _ in range(num_steps):
            self.run_one_step()
```

## 3. 运行实验并绘图

```python
def plot_results(solvers, labels):
    for solver, label in zip(solvers, labels):
        plt.plot(range(len(solver.regrets)), solver.regrets, label=label)
    plt.xlabel("Time steps")
    plt.ylabel("Cumulative regret")
    plt.title(f"{solvers[0].bandit.K}-armed bandit")
    plt.legend()
    plt.show()


np.random.seed(1)
solver = EpsilonGreedy(bandit, epsilon=0.01)
solver.run(5000)
print("epsilon=0.01 时的累积懊悔：", solver.regret)
plot_results([solver], ["epsilon=0.01"])
```
